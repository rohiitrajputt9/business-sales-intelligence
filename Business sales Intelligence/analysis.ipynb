```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("data/sales_data.xlsx")
df.head()

```

```python
df.drop_duplicates(inplace= True)
df.dropna(inplace=True)
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.head()
```

```python
df.head()
df.columns
```

```python
'''
Showing Total Revenue
'''
total_revenue = df['Total_Sales'].sum()
print("Total Revenue is : ", total_revenue)
```

```python
''' Showing Monthly Sales Trend'''
monthly_sales = df.groupby(df['Order_Date'].dt.to_period('M'))['Total_Sales'].sum()
monthly_sales
```

```python
'''Showing Top Products by Revenue'''

top_products = (
    df.groupby('Product')['Total_Sales']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
top_products
```

```python
'''Region-wise Revenue'''
region_sales = df.groupby('Region')['Total_Sales'].sum()
region_sales
```

```python
'''Category wise Revenue'''
category_sales = df.groupby('Category')['Total_Sales'].sum()
category_sales
```

```python
'''Average Order Value'''
avg_order_value = df['Total_Sales'].mean()
avg_order_value
```

```python
best_product = top_products.idxmax()
best_region = region_sales.idxmax()

print("Business Insights:")
print(f"- Top-selling product by revenue: {best_product}")
print(f"- Highest performing region: {best_region}")
```
